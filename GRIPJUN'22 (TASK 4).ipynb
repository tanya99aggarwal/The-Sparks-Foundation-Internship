```python
iyear=df1['iyear'].unique()
iyear_count=df1['iyear'].value_counts(dropna=False).sort_index()
plt.figure(figsize=(18,10))
sns.barplot(x=iyear,y=iyear_count,palette='tab10')
plt.xticks(rotation=50)
plt.xlabel('Attacking Year',fontsize=20)
plt.ylabel('Number of Attacks Each year',fontsize=20)
plt.title('Attack in Years',fontsize=30)
plt.show()
```

```python
# MAX attacks can seen in the year 2014
#MIN attacks can seen in the year1971
#A huge gype of number of attacks from 2011 to 2014
```

```python
pd.crosstab(df1.iyear,df1.region).plot(kind='area',stacked=False,figsize=(20,10))
plt.xlabel('Year',fontsize=20)
plt.ylabel('Number of Attacks',fontsize=20)
plt.title('Terrorist Activities By Region In Each Year',fontsize=25)
plt.show()
```

```python
#Region 10 has highest attacks followed by region 6
```

```python
attack=df1.country_txt.value_counts()
attack
```

```python
plt.subplots(figsize=(20,10))
sns.barplot(df1['country_txt'].value_counts()[:10].index,df1['country_txt'].value_counts()[:10].values,palette='tab10')
plt.xlabel('Countries',fontsize=20)
plt.xticks(rotation=50)
plt.ylabel('count',fontsize=20)
plt.title('Top Countries affected',fontsize=25)
plt.show()
```

```python
#Iraq had Highest number of attacks
```

```python
df=df1[['iyear','nkill']].groupby(['iyear']).sum()
fig,ax4=plt.subplots(figsize=(20,10))
df.plot(kind='bar',alpha=0.7,ax=ax4)
plt.xticks(rotation=50)
plt.xlabel('Year',fontsize=20)
plt.ylabel('Number of Killed people',fontsize=20)
plt.title('People died due to attack',fontsize=25)
top_side=ax4.spines['top']
top_side.set_visible(False)
right_side=ax4.spines['right']
right_side.set_visible(False)
```

```python
#Maximum number of peoples killed in the Year 2014, i.e mothe than 40000
```

```python
plt.subplots(figsize=(20,10))
sns.barplot(df1['city'].value_counts()[:10].index,df1['city'].value_counts()[:10].values,palette='tab10')
plt.xlabel('City',fontsize=20)
plt.xticks(rotation=50)
plt.ylabel('count',fontsize=20)
plt.title('Top Cities affected',fontsize=25)
plt.show()
```

```python
#The Unknown city is most affected ,followed by Baghdad and so on.....
```

```python
df1[["attacktype1_txt",'nkill']].groupby(['attacktype1_txt'],axis=0).sum().plot(kind='bar',figsize=(20,10),color=['pink'])
plt.xlabel('Attack type',fontsize=20)
plt.xticks(rotation=50)
plt.ylabel('Number of people',fontsize=20)
plt.title('Number of people Killed',fontsize=25)
plt.show()
```

```python
#Armed Assault and Bombing/Explosion killed the Maximum number of People
```

```python
df1[["attacktype1_txt",'nwoundus']].groupby(['attacktype1_txt'],axis=0).sum().plot(kind='bar',figsize=(20,10),color=['blue'])
plt.xlabel('Attack type',fontsize=20)
plt.xticks(rotation=50)
plt.ylabel('Number of people',fontsize=20)
plt.title('Number of Wounds',fontsize=25)
plt.show()
```

```python
#Bombing gives maximum Wounds
```

```python
plt.subplots(figsize=(20,10))
sns.countplot(df1['targtype1_txt'],order=df1['targtype1_txt'].value_counts().index,palette='gist_heat',edgecolor=sns.color_palette('mako'))
plt.xlabel('Target type',fontsize=20)
plt.xticks(rotation=90)
plt.ylabel('count',fontsize=20)
plt.title('Count of Different type Targeted people',fontsize=25)
plt.show()
```

```python
#Private citizens and property targeted the most
```

```python
df1['gname'].value_counts().to_frame().drop('Unknown').head(10).plot(kind='bar',figsize=(20,10),color=['Orange'])
plt.xlabel('Terrorist group name',fontsize=20)
plt.xticks(rotation=90)
plt.ylabel('Attack Number',fontsize=20)
plt.title('Top 10 Terrorist group attack',fontsize=25)
plt.show()
```
