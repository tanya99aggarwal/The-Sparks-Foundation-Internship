# Visualizing the Data

```python
sns.pairplot(df1)
```

```python
fig,axes=plt.subplots(1,1,figsize=(12,7))
sns.heatmap(df1.corr())
plt.show()
```

# Sub Plots

```python
fig,axes=plt.subplots(1,2,figsize=(14,5))
fig.suptitle('Total Profit VS Sale')
sns.barplot(data=df1.groupby('Sub-Category')['Sales','Profit'].agg(sum),x='Sales',y='Profit',ax=axes[1])
df1.groupby('Sub-Category')['Sales','Profit'].agg(sum).plot(kind='bar',ax=axes[0])
plt.xticks(rotation=90)
plt.show()
```

```python
fig,axes=plt.subplots(1,2,figsize=(14,5))
fig.suptitle('Total sales VS Quantity')
sns.barplot(data=df1.groupby('Sub-Category')['Sales','Quantity'].agg(sum),x='Sales',y='Quantity',ax=axes[1])
df1.groupby('Sub-Category')['Sales','Quantity'].agg(sum).plot(kind='bar',ax=axes[0])
plt.xticks(rotation=90)
plt.show()
```

```python
fig,axes=plt.subplots(1,2,figsize=(14,5))
fig.suptitle('Total Discount VS Profit')
sns.barplot(data=df1.groupby('Sub-Category')['Discount','Profit'].agg(sum),x='Discount',y='Profit',ax=axes[1])
df1.groupby('Sub-Category')['Discount','Profit'].agg(sum).plot(kind='bar',ax=axes[0])
plt.xticks(rotation=90)
plt.show()
```

```python
fig,axes=plt.subplots(1,2,figsize=(14,5))
fig.suptitle('Total Profit VS Quantity')
sns.barplot(data=df1.groupby('Sub-Category')['Profit','Quantity'].agg(sum),x='Profit',y='Quantity',ax=axes[1])
df1.groupby('Sub-Category')['Profit','Quantity'].agg(sum).plot(kind='bar',ax=axes[0])
plt.xticks(rotation=90)
plt.show()
```

# Distribution Plots

```python
fig,axes=plt.subplots(2,2,figsize=(16,8))
fig.suptitle("Distribution Plots",fontsize=16)
sns.distplot(df1['Sales'],ax=axes[0,0])
sns.distplot(df1['Profit'],ax=axes[0,1])
sns.distplot(df1['Discount'],ax=axes[1,0])
sns.distplot(df1['Quantity'],ax=axes[1,1])
plt.show()
```

```python
fig,axes=plt.subplots(2,2,figsize=(16,8))
fig.suptitle("Sales with different Shipping modes and Segments",fontsize=16)
sns.barplot(df1['Ship Mode'],df1['Sales'],ax=axes[0,0])
sns.lineplot(df1['Ship Mode'],df1['Sales'],ax=axes[0,1])
sns.barplot(df1['Segment'],df1['Sales'],ax=axes[1,0])
sns.lineplot(df1['Segment'],df1['Sales'],ax=axes[1,1])
plt.show()
```

```python
fig,axes=plt.subplots(1,1,figsize=(12,7))
sns.countplot(df1['Quantity'],hue=df1['Region'])
plt.show()
```
